```python
df['timestamp'] = pd.to_datetime(df['datetime'])

df.set_index('timestamp', inplace=True)
```

```python
grouped_data = df.resample('M').mean()
```

```python
grouped_data = grouped_data[['weather', 'temp', 'holiday', 'windspeed', 'count', 'season', 'count']]
grouped_data
```

```python
sns.lineplot(data=df, x='datetime', y='count')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Bike Counts over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
def bike_usage_by_season(df):
    
    season_counts = df.groupby('season')['count'].sum()
    
    season_labels = ['Spring', 'Summer', 'Fall', 'Winter']

    plt.bar(season_labels, season_counts)
    plt.xlabel('Season')
    plt.ylabel('Bike Usage')
    plt.title('Bike Usage by Season')
```

```python
bike_usage_by_season(df)
```
