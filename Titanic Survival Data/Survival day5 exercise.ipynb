```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

df = pd.read_csv('train.csv')

import matplotlib.pyplot as plt


survival_counts = df['Survived'].value_counts()


plt.bar(survival_counts.index, survival_counts.values)
plt.xlabel('Survived_Passengers')
plt.ylabel('Count')
plt.title('Number of Surviving/Non-Surviving Passengers')
plt.xticks(survival_counts.index, ['Non-Survived', 'Survived'])
plt.show()

```

```python

first_class_passengers = df[df['Pclass'] == 1]


total_first_class_passengers = len(first_class_passengers)


surviving_first_class_passengers = first_class_passengers[first_class_passengers['Survived'] == 1]


num_surviving_first_class_passengers = len(surviving_first_class_passengers)


proportion_surviving_first_class = num_surviving_first_class_passengers / total_first_class_passengers

print("Proportion of Surviving 1st Class Passengers: {:.2f}".format(proportion_surviving_first_class))
```

```python


grouped_counts = df.groupby(['Sex', 'Pclass']).size().unstack()

grouped_counts.plot(kind='bar', stacked=True)


plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Passenger Count by Gender and Class')


plt.legend(title='Sex')

plt.show()

```

```python

survived = df[df['Survived'] == 1]
not_survived = df[df['Survived'] == 0]


plt.hist(survived['Age'], bins=20, alpha=0.6, label='Survived', color='blue')


plt.hist(not_survived['Age'], bins=20, alpha=0.5, label='Not Survived', color='red')


plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Surviving/Non-Surviving Passengers')


plt.legend()


plt.show()
```

```python

average_age_survived = df[df['Survived'] == 1]['Age'].mean()


average_age_drowned = df[df['Survived'] == 0]['Age'].mean()

print("Average Age of Survived Passengers: {:.2f}".format(average_age_survived))
print("Average Age of Drowned Passengers: {:.2f}".format(average_age_drowned))
```

```python

mean_age = df['Age'].mean()


df['Age'] = df['Age'].fillna(mean_age)
```

```python

survival_table = df.pivot_table(index='Pclass', columns=['Sex', 'Survived'], aggfunc='size')


survival_table.columns = ['Female - Dead', 'Female - Survived', 'Male - Dead', 'Male - Survived']

print(survival_table)
```
